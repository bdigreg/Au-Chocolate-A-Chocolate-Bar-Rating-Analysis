# 0.3 Data Wrangling/Cleaning

```python
# dropping unnecessary columns
df_flavors = df_flavors.drop(columns = ['REF', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12'])
```

```python
# checking drop
df_flavors.info()
```

```python
# exploring the missing values in 'Ingredients'
df_flavors['Ingredients']
```

```python
# Split the 'Ingredients' column into two separate columns
df_flavors[['Numeric', 'String']] = df_flavors['Ingredients'].str.split('-', expand=True)

# Remove leading and trailing whitespace from both columns
df_flavors['Numeric'] = df_flavors['Numeric'].str.strip()
df_flavors['String'] = df_flavors['String'].str.strip()
```

```python
df_flavors.head()
```

```python
# dropping 'Ingredients' column because I split it into 'Numeric' and 'String'
df_flavors = df_flavors.drop(columns = ['Ingredients'])
```

```python
# checking drop column
df_flavors.head()
```

```python
# renaming 'Numeric' to 'Number_of_Ingredients' and 'String' to 'Ingredients_abbr'
df_flavors.rename(columns = {'Numeric' : 'Number_of_Ingredients', 'String' : 'Ingredients_abbr'},inplace =True)
```

```python
# checking data type of 'Number_of_Ingredients'
df_flavors['Number_of_Ingredients'].dtype
```

```python
# changing data type of 'Number_of_Ingredients' from 'object' to numerical
df_flavors['Number_of_Ingredients'] = df_flavors['Number_of_Ingredients'].astype('Float64')
```

```python
# checking data type of 'Number_of_Ingredients' again
df_flavors['Number_of_Ingredients'].dtype
```

```python
# addressing NaNs in 'Number_of_Ingredients'
df_nan = df_flavors[df_flavors['Number_of_Ingredients'].isnull() == True]
```

```python
nan_countries = df_flavors.loc[df_flavors['Company Location'].isin (['U.S.A.', 'U.K.', 'Italy', 'Canada', 'Australia', 'Belgium',
       'Japan', 'Ecuador', 'Russia', 'South Korea', 'Sao Tome & Principe',
       'Argentina', 'Germany', 'France', 'Madagascar', 'Peru', 'Spain',
       'New Zealand', 'Poland', 'Brazil','Colombia'])]
```

```python
nan_countries
```

```python
# finding stats for countries of 'Number_of_Ingredients' NaNs
nan_countries.groupby('Company Location')['Number_of_Ingredients'].agg(['min', 'max', 'mean', 'median'])
```

```python
# creating dictionary and map to replace NaN specific to each country ('Compnay Location')
replacement_values = {
    'Argentina': 4,  # Replace NaN for Argentina with 4, etc.
    'Australia': 3,  
    'Belgium': 4,
    'Brazil': 3,  
    'Canada': 3,
    'Colombia': 4,
    'Ecuador': 4,
    'France': 4,
    'Germany': 3,
    'Italy': 3,
    'Japan': 3,
    'Madagascar': 4,
    'New Zealand': 2,
    'Peru': 3,
    'Poland': 2,
    'Russia': 2,
    'Sao Tome & Principe': 3,
    'South Korea': 2,
    'Spain': 3,
    'U.K.': 3,
    'U.S.A.': 3
}
```

```python
# Define a custom function to replace NaN values
def replace_nan(row):
    location = row['Company Location']
    replacement = replacement_values.get(location)
    if pd.isna(row['Number_of_Ingredients']):
        return replacement
    else:
        return row['Number_of_Ingredients']

# Apply the custom function to replace NaN values in 'Number_of_Ingredients'
df_flavors['Number_of_Ingredients'] = df_flavors.apply(replace_nan, axis=1)
```

NaN values in the "Number_of_Ingredients" column for each company location in 'df_flavors' have been replaced with the specified values.
